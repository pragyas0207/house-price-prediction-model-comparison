# Data Correlation
#### Correlation b/w features. If its 1-->Positive correlation and if its -1--->Negative correlation

```python
plt.figure(figsize=(10,10))
# cor=data.corr()--->corr usually works only for int cols, but here all cols r taken so error
cor=data.select_dtypes(include=['number']).corr()
sns.heatmap(cor,annot=True,cmap=plt.cm.PuBu)
plt.show
```

```python
cor_target=abs(cor['Price'])
relevant_features=cor_target[cor_target>0.2]   # selecting highly correlated features
names=[index for index, value in relevant_features.items()]
names.remove('Price')
print(names)
print(len(names))

# Why selecting these highly correlated vars?
# When predicting price, we need features that r strongly related to Price
# correlaton=0--->feature has little linear relationship with Price
# correlation = +1 or -1 --->strong relationship
```

# Model Building

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
y=data.Price
features=['BuildingArea','YearBuilt','Rooms','Distance']
X=data[features]
# X=data.drop('Price',axis=1)   # all cols except Price
X_train,X_val,y_train,y_val=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
print(X_train.shape)
print(X_val.shape)
print(y_train.shape)
print(y_val.shape)
```

```python
print(X_train.columns)
print(X_train.dtypes)
print(X_train.isnull().sum())
```

```python
# Just transform means learn smtg from data
# fit_transform meand apply what was learned

from sklearn.impute import SimpleImputer
imputer=SimpleImputer()
imputed_X_train=pd.DataFrame(imputer.fit_transform(X_train))
imputed_X_val=pd.DataFrame(imputer.transform(X_val))
# After imputing u get just a numpy array not a datafarame

# to restore col names since they are lost- in the process of imputing and converting to Dataframe
imputed_X_train.columns=X_train.columns
imputed_X_val.columns=X_val.columns
```

```python
#  Train vs Validation Error
model=DecisionTreeRegressor(random_state=1)
model.fit(imputed_X_train,y_train)

# Train predictions
train_preds = model.predict(imputed_X_train)

# Validation predictions
val_preds = model.predict(imputed_X_val)

# Calculate MAE
train_mae = mean_absolute_error(y_train, train_preds)
val_mae = mean_absolute_error(y_val, val_preds)

print("Train MAE:", train_mae)
print("Validation MAE:", val_mae)
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.bar(["Train MAE", "Validation MAE"], [train_mae, val_mae])
plt.title("Train vs Validation Error")
plt.ylabel("MAE")
plt.show()
print("The model shows signs of overfitting as the training MAE (0.09) is significantly lower than the validation MAE (0.28), indicating poor generalization.â€")
print("\nThis can be fixed by limiting the depth of Decision Tree")
```

```python
# Linear Regression doesnt accept missing val cols
lr_model=LinearRegression()
lr_model.fit(imputed_X_train,y_train)
```

```python
lr_prediction=lr_model.predict(imputed_X_val)    # prediction gives a numpy array and it has no iloc func...just direct indexing
print("Actual value of house:- ",y_val.iloc[0])
print("Predicted value:- ",lr_prediction[0])



# for i in range(5):
#     print("Actual:", y_val.iloc[i], 
#           " Predicted:", prediction[i])
```

```python
from sklearn.metrics import mean_squared_error

lr_mse=mean_squared_error(y_val,lr_prediction)
print("Mean Squared Error: ",lr_mse)
rmse=np.sqrt(lr_mse)
print("Root Mean Squared Error:",rmse)

lr_mae=mean_absolute_error(y_val,lr_prediction)
print("Mean Absolute Error: ",lr_mae)
```

```python
def get_mae(y_val,pred):
    mae=mean_absolute_error(y_val,pred)
    return mae

def get_mse(y_val,pred):
    mse=mean_squared_error(y_val,pred)
    return mse
```

```python
dt_model=DecisionTreeRegressor()
dt_model.fit(imputed_X_train,y_train)

dt_prediction=dt_model.predict(imputed_X_val)
print("Actual Price of House: ",y_val.iloc[0])
print("Predicted Price of House: ",dt_prediction[0])
```

```python
dt_mse=get_mse(y_val,dt_prediction)
print("Mean Squared Error: ",dt_mse)
print("Root MSE: ",np.sqrt(dt_mse))
dt_mae=get_mae(y_val,dt_prediction)
print("Mean Absolute Error: ",dt_mae)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_model=RandomForestRegressor()
rf_model.fit(imputed_X_train,y_train)

rf_prediction=rf_model.predict(imputed_X_val)
print("Actual Price of House: ",y_val.iloc[0])
print("Predicted Price of House: ",rf_prediction[0])
```

```python
rf_mse=get_mse(y_val,rf_prediction)
print("Mean Squared Error: ",rf_mse)
print("Root MSE: ",np.sqrt(rf_mse))
rf_mae=get_mae(y_val,rf_prediction)
print("Mean Absolute Error: ",rf_mae)
```
